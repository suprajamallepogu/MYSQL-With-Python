```python
data1=data[data.index.isin(range(200,260))]
sns.boxplot(x='FirstName',y='Weight',data=data1)
plt.title("BOX PLOT")
plt.xticks(rotation=60)
plt.show()

    
```

```python
data2=data[data.index.isin(range(10,25))]
sns.lineplot(x='EmergencyContactName',y='New_Height',data=data2)
plt.title("LINE PLOT")
plt.xticks(rotation=60)
plt.show()
```

```python
data3=data[data.index.isin(range(56,71))]
sns.barplot(x='SmokingStatus',y='New_Height',data=data3)
plt.title("BAR PLOT")
plt.show()
```

```python
data4=data[data.index.isin(range(190,200))]
sns.countplot(x='ChronicConditions',data=data4)
plt.title("COUNT PLOT")
```

```python
data5=data[data.index.isin(range(200,231))]
sns.violinplot(x='Allergies',y='Weight',data = data5)
plt.title("VIOLIN PLOT")
```

```python
data6=data[data.index.isin(range(99,145))]
sns.histplot(x='DateOfLastVisit',y='New_Height',data = data6)
plt.title("Hist plot")
plt.xticks(rotation=45)
```

```python
sns.pairplot(data,hue='EmergencyContactPhone')
plt.title("Pair plot")
plt.show()
```

```python
labels = data['EmailAddress'].head(5)
sizes= data['Weight'].head(5)
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Pie chart ")
plt.show()
```
