```python
data1=data[data.index.isin(range(7,100))]
sns.boxplot(x='manufacture_country',y='cost',data=data1)
plt.show()

    
    
```

```python
data2=data[data.index.isin(range(10,25))]
sns.lineplot(x='phone_name',y='cost',data=data2)
```

```python
data3=data[data.index.isin(range(56,111))]
sns.barplot(x='charger_volts',y='cost',data=data3)
```

```python
data4=data[data.index.isin(range(190,200))]
sns.countplot(x='processor_type',data=data4)
plt.xticks(rotation=20)
```

```python
data5=data[data.index.isin(range(99,125))]
sns.violinplot(x='type',y='cost',data = data5)
plt.xticks(rotation=20)
```

```python
data6=data[data.index.isin(range(99,199))]
sns.histplot(x='phone_name',y='cost',data = data6)
plt.xticks(rotation=20)
```
